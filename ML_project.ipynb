```python
sample = df[df['ID']==	1312200]
name_list = df1['Team1Players'][0].split(', ')
name_list= name_list + df1['Team2Players'][0].split(', ')
players = []
for i in range(len(name_list)):
  if i==0:
    players.append(name_list[i][1:])
  elif i==len(name_list)-1:
    players.append(name_list[i][:len(name_list)-1])
  else:
    players.append(name_list[i])
x=[]
for i in range(len(players)):
  if sample[sample['batter']==players[i].strip("'")]['batter'].count()!=0:
    a = 4+sample[sample['batter']==players[i].strip("'")]['batsman_run'].sum()+sum(sample[sample['batter']==players[i].strip("'")]['batsman_run']==4)+(sum(sample[sample['batter']==players[i].strip("'")]['batsman_run']==6)*2)
    if sample[sample['batter']==players[i].strip("'")]['batsman_run'].sum()>=50 and sample[sample['batter']==players[i].strip("'")]['batsman_run'].sum()<100:
      a=a+8
    elif sample[sample['batter']==players[i].strip("'")]['batsman_run'].sum()>=50 and sample[sample['batter']==players[i].strip("'")]['batsman_run'].sum()<100:
      a=a+16
    elif sample[sample['batter']==players[i].strip("'")]['batsman_run'].sum()==0:
      a=a-2
  else:
    a=4
  x.append(a)
x
```
